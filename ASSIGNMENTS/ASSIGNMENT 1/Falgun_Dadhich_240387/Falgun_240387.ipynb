 Question 1: Control Flow and Loops

```python
def Fibonacci(X):
  if X <= 0:
        return []
  elif X == 1:
        return [0]
  elif X == 2:
        return [0, 1]
  d=[0,1]
  for i in range(2,X):
     d.append(d[-1] + d[-2])
  return d
N = 10
fibonacci_numbers = Fibonacci(N)
print(f"The first {N} Fibonacci numbers are: {fibonacci_numbers}")
```

Question 11: Handling Missing Values

```python
new_data = {'Age' : np.random.randint(15,40,20),
            'Salary' : np.random.randint(20000,100001,20)}
salary_df = pd.DataFrame(new_data)
nan_indices = [(1, 'Age'), (4, 'Salary'), (17, 'Salary')]
for row, col in nan_indices:
    salary_df.loc[row, col] = np.nan
print(salary_df)
```

```python
new_age = salary_df['Age'].fillna(salary_df['Age'].mean())
new_salary = salary_df['Salary'].fillna(salary_df['Salary'].mean())
salary_df['Age'] = new_age
salary_df['Salary'] = new_salary
print(salary_df)
```

Question 13: Line Plot

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
revenue = np.random.randint(100000, 500001, len(months))
plt.figure(figsize=(10, 6))
plt.plot(months, revenue, marker='o', linestyle='--', color='b', label='Monthly Revenue')
plt.title('Monthly Revenue Trend', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Revenue', fontsize=12)
plt.legend(fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Question 14: Bar Plot


```python
products = ['A', 'B', 'C', 'D', 'E']
region_1_sales = [200, 340, 290, 410, 370]
region_2_sales = [220, 300, 310, 390, 380]
x=np.arange(5)
plt.bar(x-0.125, region_1_sales,0.25, label='Region 1', color='blue')
plt.bar(x+.125, region_2_sales,0.25, label='Region 2', color='orange')
plt.title('Sales Comparison Between Regions', fontsize=16)
plt.xlabel('Products', fontsize=12)
plt.ylabel('Sales ', fontsize=12)
plt.xticks(x, products)
plt.legend(fontsize=10)

plt.show()
```
