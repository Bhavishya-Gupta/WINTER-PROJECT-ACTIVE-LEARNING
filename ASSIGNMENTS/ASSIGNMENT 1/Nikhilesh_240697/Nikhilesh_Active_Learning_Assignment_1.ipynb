Q1] Control Flow and Loops

```python
#Q1]
N=int(input("Enter a number: "))
fibo1=0
fibo2=1
fibo=fibo1+fibo2
for i in range(N):
        if i!=9:
            print(f"{fibo1},",end='')
        else:
            print(f"{fibo1}",end='')
        temp=fibo+fibo2
        fibo1=fibo2
        fibo2=fibo
        fibo=temp
```

Q4] Array Operations

```python
#Q4]
import numpy as np
array = np.random.randint(10, 101, size=25)
Max=np.max(array,axis=0)
Min=np.min(array,axis=0)
MaxIndex=np.argmax(array)
MinIndex=np.argmin(array)
print(f"The maximum value is {Max} and its index is {MaxIndex}")
print(f"The minimum value is {Min} and its index is {MinIndex}")
for i in range(25):
  if array[i]%2!=0:
    array[i]=-1
print(array)
```

Q5] Matrix Manipulation

```python
#Q5]
import numpy as np
array=np.random.randint(1,101,size=(5,5))
print("5x5 Random Matrix: ")
print(array)
diag=np.diag(array)
print("Diagonal Elements: ")
print(diag)
sum=0
for i in range(5):
  sum=sum+diag[i]
print(f"Sum of Diagonal Elements is {sum}")
```

Q6] Statistical Functions

```python
#Q6]
import numpy as np
array=np.random.rand(3,4,5)
mean=np.mean(array)
median=np.median(array)
std=np.std(array)
print(f"Mean is {mean}")
print(f"Median is {median}")
print(f"Standard Deviation is {std}")
for i in range(3):
  for j in range(4):
    for k in range(5):
      array[i][j][k]=(array[i][j][k]-mean)/std
print(array)
```

Q11] Handling Missing Values

```python
#Q11]
import pandas as pd
import numpy as np
data = {
    "Apples": np.random.randint(1, 100, 10),
    "Bananas": np.random.randint(1, 100, 10),
    "Mango": np.random.randint(1, 100, 10),
    "Guava": np.random.randint(1, 100, 10)
}
df = pd.DataFrame(data)
df.loc[3, "Apples"] = np.nan
df.loc[7, "Bananas"] = np.nan
df.loc[5, "Mangos"] = np.nan
df.loc[2, "Guava"] = np.nan
df_filled = df.copy()
df_filled = df_filled.apply(lambda col: col.fillna(col.mean()), axis=0)
print(df_filled)
```

Q12] Merging DataFrames

Q13] Line Plot

```python
#Q13]
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
Months=["January","February","March","April","May","June","July","August","September","October","November","December"]
Revenue=np.random.randint(1,100000,size=12)
array = pd.DataFrame({
    "Month": Months,
    "Revenue": Revenue
})
plt.plot(array["Month"], array["Revenue"], label="Monthly Revenue", color="red", linestyle="dashed")
plt.title("Monthly Revenue Trend", fontsize=16)
plt.xlabel("Month", fontsize=14)
plt.ylabel("Revenue", fontsize=14)
plt.legend(fontsize=12)
plt.show()
```

Q14] Bar Plot

```python
#Q14]
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
products = ["Laptop", "Phone", "Headphones", "Speaker", "Charger"]
pune_sales = np.random.randint(50, 200, size=5)
kanpur_sales = np.random.randint(50, 200, size=5)
data = pd.DataFrame({
    "Product": products,
    "Pune": pune_sales,
    "Kanpur": kanpur_sales
})
x = np.arange(len(products))
width = 0.5
plt.bar(x - width / 2, data["Pune"], width, label="Pune", color="blue")
plt.bar(x + width / 2, data["Kanpur"], width, label="Kanpur", color="yellow")
plt.title("Sales of Products in Two Regions", fontsize=16)
plt.xlabel("Products", fontsize=14)
plt.ylabel("Sales (Units)", fontsize=14)
plt.legend(fontsize=12)
plt.show()
```

Q15] Scatter Plot

```python
#Q15]
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset("iris")
sns.scatterplot(
    data=iris,
    x="petal_length",
    y="petal_width",
    hue="species",
    palette="viridis",
    style="species"
)
plt.title("Petal Length vs Petal Width by Species", fontsize=16)
plt.xlabel("Petal Length (cm)", fontsize=14)
plt.ylabel("Petal Width (cm)", fontsize=14)
plt.legend(title="Species", fontsize=12, title_fontsize=14)
plt.show()
```
