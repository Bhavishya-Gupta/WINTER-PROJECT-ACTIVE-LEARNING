Question 1

```python
def fib(n):
    if n == 0 : 
        return 0
    if n == 1 : 
        return 1
    return fib(n-1) + fib(n-2)
n = 10
i = 0
print("Fibonacci sequence upto 10 terms is : ")
while n != 0 :
    print(fib(i),end = " ")
    i =i+1
    n = n-1
```

Question 11

```python
data = {
    'A': np.random.randint(1, 100, 10),
    'B': np.random.randint(1, 100, 10),
    'C': np.random.randint(1, 100, 10)
}
df = pd.DataFrame(data)
df.loc[2, 'A'] = np.nan
df.loc[4, 'B'] = np.nan
df.loc[6, 'C'] = np.nan

print("Original DataFrame with NaN values:")
print(df)
print()

for column in df.columns:
    column_mean = df[column].mean()
    df[column] = df[column].fillna(column_mean)
print("Updated DataFrame with NaN values replaced by column mean:")
print(df)
```

```python
import matplotlib.pyplot as plt
months = ['January', 'February', 'March', 'April', 'May', 'June', 
          'July', 'August', 'September', 'October', 'November', 'December']
revenue = np.random.randint(20000, 50000, size=12)

df = pd.DataFrame({'Month' : months,'revenue' : revenue})
print("The generated dataset is : ")
print(df)
print()
plt.figure(figsize = (15,6))
plt.plot(df['Month'],df['revenue'],label = 'Revenue',linestyle = '--',color =  'r')
plt.xlabel("Months")
plt.ylabel("Revenue Generated per month")
plt.title("Revenue generated per month for one year")
plt.legend()
plt.show()
```

Question 14

```python
products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
region_1_sales = np.random.randint(100, 500, size=5)
region_2_sales = np.random.randint(100, 500, size=5)

data = pd.DataFrame({
    'Product': products,
    'Region 1': region_1_sales,
    'Region 2': region_2_sales
})

print("The assumed dataset is :")
print(data)
print()

bar_width = 0.35
x = np.arange(len(products))
plt.figure(figsize=(10, 6))

plt.bar(x - bar_width / 2, data['Region 1'], width=bar_width, label='Region 1', color='blue')
plt.bar(x + bar_width / 2, data['Region 2'], width=bar_width, label='Region 2', color='green')

plt.title('Sales Comparison of Products in Two Regions', fontsize=16)
plt.xlabel('Products', fontsize=12)
plt.ylabel('Sales', fontsize=12)
plt.xticks(x, products)
plt.legend(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```
