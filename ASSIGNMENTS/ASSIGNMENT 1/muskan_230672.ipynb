```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

num = int(input("Enter the number: "))

for i in range(num):
    print(f"F({i}) = {fibonacci(i)}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Months' : ['January', 'February', 'March', 'April', 'May', 'June', 'July',
          'August', 'September', 'October', 'November', 'December'],
    'Revenue' : np.random.randint(10000, 50000, size=12)
}
df = pd.DataFrame(data)
print(df)
plt.figure(figsize=(15, 10))
plt.plot(data['Months'], data['Revenue'], label='Monthly Revenue')
plt.title('Monthly Revenue Graph')
plt.xlabel('Month')
plt.ylabel('Revenue ($)')
plt.legend(loc='lower right')
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
Products = ['Product 1', 'Product 2', 'Product 3', 'Product 4', 'Product 5']
data =  {
    'Products' : ['Product 1', 'Product 2', 'Product 3', 'Product 4', 'Product 5'],
   'region_1_sales' : [250, 400, 300, 450, 500],
   'region_2_sales' : [300, 350, 400, 300, 450]

}
df = pd.DataFrame(data)
bar_width = 0.30
x = np.arange(len(data['Products']))
plt.bar(x - bar_width/2, data['region_1_sales'], width=bar_width, label='Region 1', color='blue')
plt.bar(x + bar_width/2, data['region_2_sales'], width=bar_width, label='Region 2', color='red')
plt.title('Sales Comparison Between Regions', fontsize=16)
plt.xlabel('Products')
plt.ylabel('Sales')
plt.xticks(x,  Products)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')


sns.scatterplot(
    data=iris,
    x='petal_length',
    y='petal_width',
    hue='species',
    palette='coolwarm',
    s=70,
    edgecolor='m'
)
plt.title('Petal Length vs Petal Width by Species')
plt.xlabel('Petal Length ', fontsize=10)
plt.ylabel('Petal Width ', fontsize=10)
plt.legend(title='Species')
plt.show()
```
