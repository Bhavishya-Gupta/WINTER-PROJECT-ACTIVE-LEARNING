**Q1**

```python
def fibo(n):
    if n == 0:
        return 0
    if n == 1:
        return 1
    x= fibo(n-1) + fibo(n-2)
    return x
   
n = 10
fib_sequence = [fibo(i) for i in range(n)]
print(fib_sequence)
```

```python
import csv

data = [
    {"Name": "Doremon", "Age": 25, "Score": 88},
    {"Name": "Nobita", "Age": 22, "Score": 72},
    {"Name": "Sezuka", "Age": 23, "Score": 91},
    {"Name": "Gian", "Age": 24, "Score": 65},
    {"Name": "Suneo", "Age": 21, "Score": 79}
]

with open('data.csv', mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=["Name", "Age", "Score"])
    writer.writeheader()
    writer.writerows(data)

with open('data.csv', mode='r') as file:
    x = csv.DictReader(file)
    for i in x:
        if int(i["Score"]) > 75:
            print(i)
```

```python
import numpy as np

arr = np.random.random((3, 4, 5))

print("Array:\n", arr)
mean = np.mean(arr)
median = np.median(arr)
std_dev = np.std(arr)

print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
normal_arr = (arr - mean) / std_dev

print("Normal Array:\n", normal_arr)
```

**Q13**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42) 
months = ['January', 'February', 'March', 'April', 'May', 'June', 
          'July', 'August', 'September', 'October', 'November', 'December']
revenue = np.random.randint(10000, 50000, size=12)

data = {'Month': months, 'Revenue': revenue}
df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))
plt.plot(df['Month'], df['Revenue'], marker='o', linestyle='-', color='b', label='Monthly Revenue')

plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue (in USD)')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

**Q14**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
sales_region_1 = np.random.randint(20, 100, size=5)
sales_region_2 = np.random.randint(20, 100, size=5)

data = {
    'Product': products,
    'Region 1': sales_region_1,
    'Region 2': sales_region_2
}
df = pd.DataFrame(data)


bar_width = 0.35
index = np.arange(len(products))

fig, ax = plt.subplots(figsize=(10, 6))
bar1 = ax.bar(index, df['Region 1'], bar_width, label='Region 1')
bar2 = ax.bar(index + bar_width, df['Region 2'], bar_width, label='Region 2')


ax.set_xlabel('Product')
ax.set_ylabel('Sales')
ax.set_title('Sales Comparison Between Regions')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(products)
ax.legend()

# Step 4: Display the plot
plt.tight_layout()
plt.show()
```
