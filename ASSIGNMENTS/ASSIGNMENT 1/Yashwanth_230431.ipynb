```python

def fib(num):
    if num == 0:
        return 0
    elif num == 1:
        return 1
    else:
        return fib(num-2)+fib(num-1)
 
 
n =10;
for i in range(0, n):
    print(fib(i), end=" ")
```

```python
import csv

data = [
    ["Name", "Age", "Score"],
    ["seeta", 20, 20],
    ["geeta", 21, 49],
    ["ram", 22, 79],
    ["shyam", 22, 99],
    ["ghanshyam", 18, 85]
]

dataFile = 'data.csv'
with open(dataFile, mode="w", newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data)

with open(dataFile, mode='r') as file:
    reader = csv.reader(file)
    next(reader) 
    for row in reader:
        if int(row[2]) > 75:
            print(row)

```

```python
import matplotlib.pyplot as plt
import numpy as np

months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
revenue = np.random.randint(1000, 10000, size=12)

plt.figure(figsize=(13, 6))
plt.plot(months, revenue, label='Revenue', color='blue', marker='o')

plt.title('Monthly Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.legend()

plt.grid(True)
plt.show()
```

Question 14

```python
import matplotlib.pyplot as plt
import numpy as np

products = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
region_1_sales = [150, 200, 250, 300, 350]
region_2_sales = [100, 150, 200, 250, 300]

bar_width = 0.4
index = np.arange(len(products))

fig, ax = plt.subplots(figsize=(10, 6))
bar1 = ax.bar(index, region_1_sales, bar_width, label='Region 1')
bar2 = ax.bar(index + bar_width, region_2_sales, bar_width, label='Region 2')

ax.set_title('Sales Comparison Between Regions for Each Product')
ax.set_xlabel('Products')
ax.set_ylabel('Sales')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(products)
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', palette='Set1', s=100)

plt.title('Petal Length vs Petal Width for Iris Species', fontsize=14)
plt.xlabel('Petal Length ', fontsize=12)
plt.ylabel('Petal Width ', fontsize=12)

plt.show()
```
